```python
# Remove rows where 'TipeFaskes' is 'Dokter Gigi'
data_filtered = data[data['TipeFaskes'] != 'Dokter Gigi']

# Display the filtered dataset to confirm
data_filtered
```

```python
# Check unique values in the 'TipeFaskes' column after filtering
unique_faskes_types_filtered = data_filtered['TipeFaskes'].unique()
unique_faskes_types_filtered
```

```python
# Define the function to filter data
def filter_by_faskes_type_and_province(data, faskes_type=None, province=None):
    filtered_data = data.copy()
    
    # Filter by TipeFaskes if input is provided
    if faskes_type:
        filtered_data = filtered_data[filtered_data['TipeFaskes'].str.contains(faskes_type, case=False, na=False)]
    
    # Filter by Provinsi if input is provided
    if province:
        filtered_data = filtered_data[filtered_data['Provinsi'].str.contains(province, case=False, na=False)]
    
    return filtered_data
```

```python
# Simulasi input pengguna
province = input("Masukkan provinsi kamu (contoh: Nanggroe Aceh Darussalam): ").strip()
faskes_type = input("Masukkan tipe fasilitas kesehatan (contoh: Rumah Sakit, Puskesmas, Klinik): ").strip()

# Panggil fungsi filter
filtered_results = filter_by_faskes_type_and_province(
    data=data_filtered,
    faskes_type=faskes_type if faskes_type else None,
    province=province if province else None
)

# Tampilkan hasil
if not filtered_results.empty:
    print("Berikut adalah fasilitas kesehatan yang sesuai dengan kriteria Anda:")
    display(filtered_results)
else:
    print("Maaf, tidak ditemukan fasilitas kesehatan yang sesuai dengan kriteria.")
```
